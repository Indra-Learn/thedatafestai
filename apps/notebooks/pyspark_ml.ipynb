```python
import pandas as pd
import requests
from io import BytesIO
from kaggle.api.kaggle_api_extended import KaggleApi
import zipfile


def public_api_request(base_url: str, api_url: str, headers: dict=None, method: str='get', params: dict={}):
    full_url = base_url + "/" + api_url
    out = dict()
    if headers is None:
        headers = {"content-type": "application/json; charset=utf-8"}
    try:
        if method.lower() == "get":
            response = requests.get(full_url, headers=headers, params=params)
            if response.status_code == 200:
                out = response.json()
            else:
                raise Exception(f"Not a valid response from {'sample_url'}")
    except Exception as e:
        print(f"Error: {e}")
    else:
        return out
```

### Fetch Reviews from [Carwale](https://www.carwale.com/land-rover-cars/range-rover/reviews/): 

```python
# https://www.carwale.com/api/userreviews/listingpage/?m=1&isApiCall=true&page=1&carname=range-rover&makename=land-rover&ratingCriteria=0&sortingCriteria=1
# https://www.carwale.com/api/userreviews/listingpage/?m=1&isApiCall=true&page=1&carname=range-rover-2018-2022&makename=land-rover&ratingCriteria=0&sortingCriteria=1
carwale_base_url = "https://www.carwale.com/"
carwale_user_reviews_api_url = "api/userreviews/listingpage/?"
carwale_user_reviews_api_url_params = {
    "m": 1,
    "isApiCall": "true",
    "page": 1,
    "carname": "range-rover", 
    "makename": "land-rover", 
    "ratingCriteria": 0, 
    "sortingCriteria": 1
}
carname_list = ["range-rover", "range-rover-2018-2022", "range-rover-2014-2018", 
                "range-rover-pre-2009", "range-rover-2013-2014", "range-rover-2012-2013",
                "range-rover-2010-2012", "range-rover-2009-2010"]

# out = public_api_request(base_url=carwale_base_url, 
#                         api_url=carwale_user_reviews_api_url,
#                         method="get",
#                         params=carwale_user_reviews_api_url_params)
# print([i for i in out.keys()])

# expertReviews_df = pd.DataFrame(out.get("expertReviews"))
# expertReviews_df.loc[:, ["title", "description", "categoryMaskingName"]]

# ReviewsWithImages_df = pd.DataFrame(out.get("userReviewList").get("ReviewsWithImages"))
# ReviewsWithImages_df.loc[:, ["Title", "Description", "RatingInfo", "VehicleInfo", "Upvotes", "Downvotes"]].head()

all_userReviewList_df = pd.DataFrame({"car_name":[], "Title":[], "Description":[], "RatingInfo":[], "VehicleInfo":[], "Upvotes":[], "Downvotes":[]})
for index, car_name in enumerate(carname_list):
    carwale_user_reviews_api_url_params["carname"] = car_name
    try:
        # print(f"loop no: {index}")
        out = public_api_request(base_url=carwale_base_url, 
                                api_url=carwale_user_reviews_api_url,
                                params=carwale_user_reviews_api_url_params)
        if len(out.get("userReviewList").get("Reviews")) > 0:
            userReviewList_df = pd.DataFrame(out.get("userReviewList").get("Reviews"))
            userReviewList_df["car_name"] = car_name
    except Exception as e:
        print(f"error happened in loop-{index}: {e}")
    else:
        all_userReviewList_df = pd.concat([all_userReviewList_df, 
                                        userReviewList_df.loc[:, ["car_name", "Title", "Description", "RatingInfo", "VehicleInfo", "Upvotes", "Downvotes"]]],
                                        ignore_index=True)
print(f"size of all_userReviewList_df: {all_userReviewList_df.shape}")
all_userReviewList_df.head(5)
```
